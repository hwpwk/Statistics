```python
%matplotlib inline

import pandas as pd
```

```python
# xlsxファイルを扱う
# xlrdのインストール
# 参考:https://note.nkmk.me/python-pandas-read-excel/
```

```python
# 複数のシートがある場合、最初のシートのみが読み込まれる
df = pd.read_excel('nissan_stadium.xlsx')
```

```python
df
```

```python
# 必要なカラムのみ抽出
df2 = df[['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 7']]
```

```python
df2
```

```python
df3 = df2.iloc[12: , :]
```

```python
df3
```

```python
df3.columns
```

```python
# カラム名の変更
# 参考:https://to-kei.net/python/data-analysis/pandas-index-columns/
df3.columns = ['年度(平成)', '総数', 'サッカー', '陸上競技', 'イベント/その他']
```

```python
df3
```

```python
df4 = df3.drop([12, 13])
```

```python
df4
```

```python
# 「平成21年度」の「平成」を取り除く
# 参考:https://note.nkmk.me/python-pandas-replace/
df5 = df4.replace('平成21年度', '21年度')
```

```python
df5
```

```python
# plot()メソッドを呼ぶと、数値の列すべてがプロットされ、indexがx軸として使われる
# よって、「年度(平成)」をindexとして割り当てることにする
# set_index()メソッドを使うとpandas.DataFrameの既存の列をindex（行名、行ラベル）に割り当てることができる
# 参考:https://note.nkmk.me/python-pandas-set-index/
df6 = df5.set_index('年度(平成)')
```

```python
df6
```

```python
df6.plot.bar(figsize=(10,10))
```

```python
df6.plot.line(figsize=(10,10))
```


陸上競技の総数が徐々に上がっている（平成29年度はおそらく途中までの集計にもかかわらず過去最高）

```python
# 競技毎の総数に対する割合を算出したカラムを追加
df6['サッカー割合'] = df6['サッカー'] / df6['総数']
df6['陸上競技割合'] = df6['陸上競技'] / df6['総数']
df6['イベント/その他割合'] = df6['イベント/その他'] / df6['総数']
```

```python
df6
```

```python
# 積み上げ棒グラフ:stacked=True
# 参考URL:https://note.nkmk.me/python-pandas-plot/
df6[['サッカー割合', '陸上競技割合', 'イベント/その他割合']].plot.bar(stacked=True, figsize=(10,10))
```

```python
# 折れ線グラフ
# 参考:https://note.nkmk.me/python-pandas-plot/
df6[['サッカー割合', '陸上競技割合', 'イベント/その他割合']].plot.line(figsize=(10,10))
```

陸上競技の割合が上昇傾向
サッカーの割合については、上昇、下降どちらも確認でき、傾向にトレンドは見られない
